# 도표 5-9. PSI 데이터 업로드의 파이썬 코드

```python
import pandas as pd

# 엑셀 파일에서 데이터를 읽어 온다.
file_path = '/content/sample_data/PSI_DATA.xlsx'
data = pd.read_excel(file_path)

data.head()
```

# 도표 5-10. 꺾은 선형 챠트의 파이썬 코드

```python
# matplotlib.pyplot을 plt로 축약하여 가져온다. 데이터를 그래픽으로 표현하는 데 사용된다.
import matplotlib.pyplot as plt

# 그래프의 크기를 가로 10인치, 세로 6인치로 설정한다.
plt.figure(figsize=(10, 6))

# 'Week' 데이터를 x축으로, 'P', 'S', 'I' 각각을 y축 데이터로 하여 선 그래프를 그린다.
# 각 선에 레이블을 달아 범례에 표시되게 하며, 각각 다른 색상과 마커('o'는 원형 마커)를 지정한다.
plt.plot(data['Week'], data['P'], label='Production (P)', color='blue', marker='o')
plt.plot(data['Week'], data['S'], label='Sales (S)', color='red', marker='o')
plt.plot(data['Week'], data['I'], label='Inventory (I)', color='green', marker='o')

# 그래프의 제목을 'Weekly PSI Trend'로 설정한다.
plt.title('Weekly PSI Trend')
# x축과 y축에 각각 'Week'과 'Value'라는 레이블을 부여하다.
plt.xlabel('Week')
plt.ylabel('Value')

# y축의 값을 천 단위 구분 기호(콤마)를 사용하여 포맷팅한다. 예를 들어, 1000은 '1,000'으로 표시된다.
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# 그래프의 범례를 표시한다. label 파라미터로 설정된 내용이 범례로 나타난다.
plt.legend()

# 그래프에 그리드를 추가한다. 이는 데이터의 위치를 더 명확하게 파악할 수 있도록 도와준다.
plt.grid(True)

# 모든 설정을 완료한 후 그래프를 화면에 표시한다.
plt.show()
```

# 도표 5-12. PSI 혼합형 차트의 파이썬 코드

```python
# matplotlib.pyplot을 plt로 축약하여 가져옵니다. 이 라이브러리는 데이터를 그래픽으로 표현하는 데 사용됩니다.
import matplotlib.pyplot as plt
# matplotlib의 ticker 모듈을 불러옵니다. 이 모듈은 축의 틱 포매팅에 사용됩니다.
import matplotlib.ticker as ticker

# 그래프의 크기를 설정합니다 (너비 12인치, 높이 6인치).
plt.figure(figsize=(12, 6))

# 생산량(P)과 판매량(S) 데이터를 선 그래프로 표현합니다.
# 'Week' 열을 x축으로, 'P' 열을 y축으로 하여 파란색 선으로 그립니다.
plt.plot(data['Week'], data['P'], label='Production (P)', color='blue')
# 'Week' 열을 x축으로, 'S' 열을 y축으로 하여 빨간색 선으로 그립니다.
plt.plot(data['Week'], data['S'], label='Sales (S)', color='red')

# 재고량(I) 데이터를 막대 그래프로 표현합니다.
# 'Week' 열을 x축으로, 'I' 열을 y축으로 하여 녹색 막대로 그립니다.
# 막대의 투명도는 0.6, 너비는 0.4로 설정합니다.
plt.bar(data['Week'], data['I'], label='Inventory (I)', color='green', alpha=0.6, width=0.4)

# 그래프의 제목과 x축, y축에 레이블을 추가합니다.
plt.title('Weekly PSI Trend')
plt.xlabel('Week')
plt.ylabel('Values')

# y축의 값에 천 단위 구분자(쉼표)를 포맷터를 사용하여 추가합니다.
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# 범례를 추가합니다. 범례는 각 데이터 세트를 설명하는 데 사용됩니다.
plt.legend()

# 그리드를 표시합니다. 이는 데이터 포인트의 정확한 위치를 파악하기 쉽게 도와줍니다.
plt.grid(True)

# 설정한 모든 옵션으로 그래프를 표시합니다.
plt.show()
```

# 도표 5-16. Potly로 구현한 영역형 차트의 파이썬 코드

```python
# Seaborn은 매트플롯라이브(matplotlib) 기반의 고급 시각화 라이브러리입니다.
# 통계적 그래프를 쉽고 아름답게 그릴 수 있도록 설계되었습니다.
import seaborn as sns
# Plotly Express는 Plotly의 고수준 인터페이스로, 복잡한 그래프를 간단한 명령어로 빠르게 생성할 수 있게 해줍니다.
# 인터랙티브한 그래프를 지원합니다.
import plotly.express as px

# Seaborn 라이브러리를 사용하여 데이터의 'I' 값을 'Week'에 따라 선 그래프로 표현합니다.
plt.figure(figsize=(10, 6))  # 그래프의 크기를 10x6 인치로 설정합니다.
# 'Week'를 x축으로, 'I'를 y축으로 하는 선 그래프를 그립니다.
sns.lineplot(data=data, x='Week', y='I', color='Slateblue')
plt.fill_between(data['Week'], data['I'], color="skyblue", alpha=0.4)  # 선 아래의 영역을 하늘색으로 채웁니다.
plt.title('Inventory Levels Over Weeks (Seaborn)')  # 그래프의 제목을 설정합니다.
plt.xlabel('Week')  # x축의 레이블을 'Week'로 설정합니다.
plt.ylabel('Inventory (I)')  # y축의 레이블을 'Inventory (I)'로 설정합니다.
plt.grid(True)  # 그리드를 표시하여 데이터의 위치를 더 쉽게 파악할 수 있게 합니다.

# 데이터 포인트마다 데이터 레이블을 추가합니다.
for i, txt in enumerate(data['I']):  # 'I' 열의 각 값과 해당 인덱스에 대해 반복합니다.
    # 각 데이터 포인트에 레이블을 추가합니다.
    plt.annotate(f'{txt:.2f}', (data['Week'][i], data['I'][i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=10)

plt.show()  # 위에서 설정한 그래프를 표시합니다.

# Plotly 라이브러리를 사용하여 인터랙티브한 영역 그래프를 생성합니다.
# 'Week'를 x축으로, 'I'를 y축으로 사용하여 영역 그래프를 생성하고 레이블을 설정합니다.
fig = px.area(data, x='Week', y='I', title='Inventory Levels Over Weeks (Plotly)', labels={'I': 'Inventory (I)'})
fig.show()  # 생성된 그래프를 표시합니다.

```

```python
import pandas as pd
import numpy as np

# Provided data for WH, TRANSPORTATION_COST, INBOUND_COST, and WH_COST
data = {
    'WH': list(range(1, 11)),
    'TRANSPORTATION_COST': [
        54.49321321, 21.28860582, 10.77669804, 8.10010903, 7.388145916,
        7.0575149, 6.941968819, 6.061868007, 5.294103579, 5.149157417
    ],
    'INBOUND_COST': [
        41.69837795, 30.27093705, 32.92507523, 34.14210727, 34.64971511,
        34.72368761, 34.82347284, 35.44452534, 35.41805534, 35.35611841
    ],
    'WH_COST': [
        4.991211585, 4.582853021, 3.827908843, 3.748599598, 3.759652864,
        3.727361851, 3.72648425, 3.696875621, 3.623403075, 3.610311073
    ]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Calculations for D, E, 비율D, 비율E
# Base inventory percentage at WH=1
base_inventory = 100

# Compute the Inventory Growth Multiplier (Square Root Law)
df['Inventory_Percent'] = base_inventory * np.sqrt(df['WH'])

# Calculate D = WH_COST * Inventory_Percent / 100
df['D'] = df['WH_COST'] * df['Inventory_Percent'] / 100

# Calculate E = TRANSPORTATION_COST + INBOUND_COST + D
df['E'] = df['TRANSPORTATION_COST'] + df['INBOUND_COST'] + df['D']

# Calculate 비율D and 비율E based on the first row values of D and E
base_D = df.loc[0, 'D']
base_E = df.loc[0, 'E']

df['비율D'] = (df['D'] / base_D) * 100
df['비율E'] = (df['E'] / base_E) * 100

# Selecting the desired columns to display
result_df = df[['WH', 'D', '비율D', 'E', '비율E']]
result_df.round(2)
```

가상의 기업 KR, SY 그리고 BM의 판매가격 하락과 SCM 운영 수준을 분석하고자한다. 판가하락_기준정보01 파일을 읽고, 가상의 기업별 계산한 필드의 계산결과를 테이블 형태로 보여주기 바란다. 계산과정은 한글로 표시하기 바란다. 금액은 소수점 첫째자리까지만 표시한다. 비율은 소수점 둘째자리까지만 표시한다. 계산할 필드에 대한 설명은 아래와 같다. 일 판매가격 하락율(%/일)은 년 판매가격 하락율(%/년)/365와 리드타임(일)을 곱해서 계산한다. 재고 이자율(%)은 리드타임(일)과 이자율(%/년)/365을 곱해서 계산한다. 판매가격 하락 손실(억원/년)은 일 목표 매출액(억원/일)에 일 판매가격 하락율(%/일)/100과 365을 곱해서 계산한다. 재고금액(억원)은 평균매출원가(억원/일)에 리드타임(일)을 곱해서 계산한다. 재고 자본비용(억원/년)은 재고금액(억원)에 이자율(%/년)/100을 곱해서 계산한다. 총 기회비용(억원/년)은 판매가격 하락 손실(억원/년)과 재고 자본비용(억원/년)을 더하여 계산한다. 재고 유지비용(억원/년)은 재고금액(억원)에 재고 유지비용 비율(%)/100을 곱해서 계산한다. 영업이익 중 낭비 비율(%)은 총 기회비용(억원/년)과 재고 유지비용(억원/년) 우선 더한 후에 년간 목표 영업이익(억원/년)을 나누고 100을 곱해서 계산한다.

```python
import pandas as pd

# Load the provided Excel file
file_path = '/content/sample_data/판가하락_기준정보01.xlsx'
df = pd.read_excel(file_path)

df.head()
```

```python
# Calculating the required fields
df['일 판매가격 하락율(%/일)'] = df['년 판매가격 하략율(%/년)'] / 365 * df['리드타임(일)']
df['재고 이자율(%)'] = df['리드타임(일)'] * df['이자율(%/년)'] / 365
df['판매가격 하락 손실(억원/년)'] = df['일 목표 매출액(억원/일)'] * df['일 판매가격 하락율(%/일)'] / 100 * 365
df['재고금액(억원)'] = df['평균매출원가(억원/일)'] * df['리드타임(일)']
df['재고 자본비용(억원/년)'] = df['재고금액(억원)'] * df['이자율(%/년)'] / 100
df['총 기회비용(억원/년)'] = df['판매가격 하락 손실(억원/년)'] + df['재고 자본비용(억원/년)']
df['재고 유지비용(억원/년)'] = df['재고금액(억원)'] * df['재고 유지비용 비율(%)'] / 100
df['영업이익 중 낭비 비율(%)'] = (df['총 기회비용(억원/년)'] + df['재고 유지비용(억원/년)']) / df['년간 목표 영업이익(억원/년)'] * 100

# Formatting the values to the required precision
df_formatted = df.round({'일 판매가격 하락율(%/일)': 1, '재고 이자율(%)': 1, '판매가격 하락 손실(억원/년)': 1,
                         '재고금액(억원)': 1, '재고 자본비용(억원/년)': 1, '총 기회비용(억원/년)': 1,
                         '재고 유지비용(억원/년)': 1, '영업이익 중 낭비 비율(%)': 2})

df_formatted[['가상의 기업', '일 판매가격 하락율(%/일)', '재고 이자율(%)', '판매가격 하락 손실(억원/년)',
              '재고금액(억원)', '재고 자본비용(억원/년)', '총 기회비용(억원/년)', '재고 유지비용(억원/년)',
              '영업이익 중 낭비 비율(%)']]
```

```python
# Save the formatted dataframe to an Excel file
output_file_path = '/content/sample_data/재고현황_분석결과.xlsx'
df_formatted.to_excel(output_file_path, index=False)

output_file_path
```
